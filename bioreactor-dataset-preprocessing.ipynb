## We know from the documentation that the dataset contains 100 Batches: 

Therefore, we will locate the different batches, subset the data for the different batches.

See the batches are grouped based upon what control strategy was used with different batch lenghts etc.,

Batches 1-30: Controlled by recipe driven approach  
Batches 31-60: Controlled by operators  
Batches 61:90: Controlled by an Advanced Process Control (APC) solution using the Raman spectroscopy  
Batches 91: 100 Contain faults resulting in process deviations.  

```python
# find the indices of the batch start (i.e., where time  == 0.2)

batch_start_indices = df['Time (h)'].index[df['Time (h)'] == 0.2]

print(batch_start_indices)
```

```python
# find the indices of the batch end (i.e., where the time detla from )

def find_batch_end_indices(df, time_col='Time (h)', reset_value=0.2):
    """Find indices where batches end."""
    # Find indices where time resets
    reset_indices = df[df[time_col] == reset_value].index

    # Take the row before each reset as batch end
    batch_end_indices = reset_indices - 1

    # Filter out negative indices (just in case)
    batch_end_indices = batch_end_indices[batch_end_indices >= 0]

    # Include last row as end of final batch
    batch_end_indices = batch_end_indices.append(pd.Index([df.index[-1]]))

    return batch_end_indices

# Call the function
batch_end_indices = find_batch_end_indices(df)
print(batch_end_indices)
```

```python
# Initialize a new column
df['batch_id'] = 0

# loop through the rowsa enumerating each batch with an ID from the start and end
for i, (start, end) in enumerate(zip(batch_start_indices, batch_end_indices), start=1):
    df.loc[start:end, 'batch_id'] = i
```

```python
# check the new column
df.tail(10)
```

```python
## save a subset of the data
subset = df[(df['batch_id'] >= 1) & (df['batch_id'] <= 10)]
subset.to_csv('batch-1-10.csv', index=False)
```
